```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def learning_curve(path,section_from=0):
    df = pd.read_csv(path)
    df[section_from:].loss.plot()
    df[section_from:].val_loss.plot()
    plt.legend(['TrainingLoss', 'ValidationLoss'], loc='upper right')
```

```python
learning_curve("log.csv",2)
```
